```python
# Logic Building Task – 2
## Data Science Internship – February 2026
### Name: P Navya Kinnera
### GitHub: Navyakinnera19


sentence = "python is easy and python is powerful"

words = sentence.split()
unique_words = set(words)

print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)




employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
print("\nHighest Salary:", highest_employee, "-", employees[highest_employee])




numbers = [45, 22, 89, 10, 66]

max_value = numbers[0]
min_value = numbers[0]

for num in numbers:
    if num > max_value:
        max_value = num
    if num < min_value:
        min_value = num

print("\nList:", numbers)
print("Max:", max_value)
print("Min:", min_value)




prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print("\nProducts above 1000:", count)



attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)

attendance_percentage = (present_count / total_days) * 100

print("\nAttendance Percentage:", attendance_percentage)




phone_numbers = [9876543210, 9123456789, 9876543210]

unique_numbers = set(phone_numbers)

print("\nUnique phone numbers:", unique_numbers)


text = "python"

frequency = {}

for char in text:
    if char in frequency:
        frequency[char] += 1
    else:
        frequency[char] = 1

print("\nCharacter Frequency:", frequency)




sample_list = [10, 20, 30]

converted_tuple = tuple(sample_list)

print("\nTuple:", converted_tuple)




employee_data = {"Ravi": 75000, "Anita": 68000}

if "Ravi" in employee_data:
    print("\nEmployee exists")
else:
    print("\nEmployee does not exist")




marks = [70, 80, 65, 75]

average = sum(marks) / len(marks)

print("\nAverage Marks:", average)
```
